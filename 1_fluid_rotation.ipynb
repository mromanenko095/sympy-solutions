# Вращение жидкости

В сосуд, имеющий форму прямого кругового цилиндра, налита жидкость, например вода.
Сосуд вращается с постоянной угловой скоростью ω относительно оси цилиндра.
Требуется определить, какую форму примет поверхность жидкости, если вращение продолжается достаточно долго.
При построении модели мы будем предполагать, что сосуд достаточно широкий и глубокий, это позволит пренебречь разными поверхностными эффектами около боковых стенок сосуда.

Искомая поверхность должна быть поверхностью вращения. Для нахождения ее формы достаточно рассмотреть осевое сечение сосуда и найти форму соответствующей кривой, из которой мы потом сможем сформировать саму поверхность.

````python
from sympy import *
from IPython.display import display
````

С помощью команды ```symbols``` создадим необходимые переменные. Параметр $g$, представляющий собой ускорение свободного падения - всегда строго положительный.

````python
x, omega, y0, C1 = symbols("x omega y0 C1")
g = symbols("g", positive=True)
display(x, omega, y0, C1, g)
````

Уравнение $$y' = \frac{\omega^2x}{g}$$ является простейшим и решается непосредственным интегрированием его правой части. Следовательно, для решения уравнения с помощью SymPy необходимо определить только правую часть уравнения - запишем ее в переменную ```ode_rhs```.

````python
ode_rhs = omega ** 2 * x / g
display(ode_rhs)
````

Общее решение простейшего дифференциального уравнения $$y' = f(x)$$ - неопределенный интеграл от правой части $f(x)$ этого уравнения. Неопределенные интегралы вычисляем с помощью команды ```integrate(f, x)```, где $f$ - подынтегральное выражение, $x$ - переменная интегрирования. Константа интегрирования $C_1$ автоматически к интегралу не прибавляется - это необходимо сделать вручную. Сохраним общее решение дифференциального уравнения в переменной ```dsol```.

````python
dsol = integrate(ode_rhs, x) + C1
display(dsol)
````
